```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")

rs = classifier("What a bullshit guy")

print(rs)
```

```python
generator = pipeline("text-generation",  model="distilgpt2")
```

```python
res = generator('In this course i will teach you', max_length=30, num_return_sequences=2)

# num_return_sequences(int, optional, defaults to 1) — The number of independently computed returned sequences for each element in the batch
# max_length (int, optional, defaults to 20) — The maximum length the generated tokens can have. Corresponds to the length of the input prompt + max_new_tokens. Its effect is overridden by max_new_tokens, if also set
# https://huggingface.co/docs/transformers/main_classes/text_generation

print(res)
```

```python
classifier = pipeline("zero-shot-classification")
```

```python
res = classifier('Babasaheb Ambedkar was a great man',
                 candidate_labels=['education', 'politics', 'business', 'history']
                 )
print(res)
```

```python
from transformers import AutoTokenizer , AutoModelForSequenceClassification
```

```python
classifier = pipeline("sentiment-analysis")
res = classifier("Be Educated, Be Organised & Be Agitated")
print(res)

model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer =  AutoTokenizer.from_pretrained(model_name)

classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
res = classifier("Be Educated, Be Organised & Be Agitated")
print(res)
```

```python
sequence = "Babasaheb ambedkar was a great man"
res = tokenizer(sequence)
print(res)

tokens = tokenizer.tokenize(sequence)
print(tokens)

ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)

decoded_string = tokenizer.decode(ids)
print(decoded_string)
```
